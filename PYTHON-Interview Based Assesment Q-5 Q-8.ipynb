```python
import requests
import json
import csv
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings

warnings.filterwarnings("ignore")

%matplotlib inline
```

```python
from bs4 import BeautifulSoup
# API URL
url = "http://api.tvmaze.com/singlesearch/shows?q=westworld&embed=episodes"
# Send GET request to the API
response = requests.get(url)
# Check if the request was successful
if response.status_code == 200:
    # Extract JSON data from the response
    data = response.json()
    # Extract show information
    show_id = data['id']
    show_url = data['url']
    show_name = data['name']
    
    # Extract episode information
    episodes = data['_embedded']['episodes']
    episodes_data = []
    for episode in episodes:
        episode_id = episode['id']
        episode_season = episode['season']
        episode_number = episode['number']
        episode_type = episode['type']
        episode_airdate = episode['airdate']
        episode_airtime = episode['airtime']
        episode_runtime = episode['runtime']
        episode_rating = episode['rating']['average']
        episode_summary = BeautifulSoup(episode['summary'], "html.parser").get_text()
        episode_medium_image = episode['image']['medium']
        episode_original_image = episode['image']['original']
        
        episodes_data.append({
            'id': episode_id,
            'url': show_url,
            'name': show_name,
            'season': episode_season,
            'number': episode_number,
            'type': episode_type,
            'airdate': episode_airdate,
            'airtime': episode_airtime,
            'runtime': episode_runtime,
            'average rating': episode_rating,
            'summary': episode_summary,
            'medium image link': episode_medium_image,
            'original image link': episode_original_image
        })

    # Create a DataFrame for episode information
    df_episodes = pd.DataFrame(episodes_data)
    # Print the DataFrame
    print("Episode Information:")
    print(df_episodes)
else:
    print("Failed to retrieve data. Status code:", response.status_code)
```

### Get all the overall ratings for each season and using plots compare the ratings for all the seasons, like season 1 ratings, season 2, and so on.

```python
season_rating=dfc.groupby('season')['average rating'].mean()
plt.figure(figsize=(10, 6))
season_rating.plot(kind='bar', color='blue')
plt.xlabel('Season')
plt.ylabel('Average Rating')
plt.title('Ratings by Season')
plt.xticks(rotation=45)
plt.show()
```

### Get all the episode names, whose average rating is more than 8 for every season

```python
dfc_rate=dfc[dfc['average rating'] >8]
sns.barplot(x='season', y='average rating', data=dfc_rate)
plt.xlabel('Season')
plt.ylabel('Rating')
plt.title('Episodes with Average Rating greater than 8')
plt.show()
```

```python
# Convert the 'airdate' column to datetime
dfc['airdate'] = pd.to_datetime(dfc['airdate'])

# Filter episodes aired before May 2019
dfc_before_2019 = dfc[dfc['airdate'] < pd.to_datetime('2019-05-01')]

# Get the episode names
episode_names = dfc_before_2019['name']
sns.countplot(data=dfc_before_2019, x='season')

# Set the plot title and labels
plt.title('Episodes Aired Before May 2019')
plt.xlabel('Season')
plt.ylabel('Count')

# Show the plot
plt.show()
```

### Get the episode name from each season with the highest and lowest rating



```python
low=dfc.groupby('season')['average rating'].min() # Group the data by season and find the episode with the lowest rating in each season
high=dfc.groupby('season')['average rating'].max()# Group the data by season and find the episode with the highest rating in each season
```

```python
plt.figure(figsize=(10,6))
plt.subplot(1,2,1)
low.plot(kind='bar')
plt.xlabel('Season')
plt.ylabel('lowest Rating')
plt.title('Season with lowest rating')

plt.subplot(1,2,2)
high.plot(kind='bar')
plt.xlabel('Season')
plt.ylabel('highest Rating')
plt.title('season with highest rating')

plt.suptitle('Season with Highest and lowest ratings')
```

### Get the summary for the most popular ( ratings ) episode in every season

```python
most_popular_episodes = dfc.groupby('season')['average rating'].idxmax()# Group the data by season and find the episode with the highest rating in each season
# Get the summary for the most popular episodes
summary_most_popular = dfc.loc[most_popular_episodes, ['season', 'summary']]
# Print the summary for the most popular episodes
for index, row in summary_most_popular.iterrows():
    season = row['season']
    summary = row['summary']
    print(f"Season {season} - Summary: {summary}")
```
